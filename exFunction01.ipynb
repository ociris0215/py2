```python
import matplotlib.pyplot as plt
# Data to plot
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Create a figure and an axis
fig, ax = plt.subplots()

# Plot the data
ax.plot(x, y)

# Add a title and labels
ax.set_title("Line Plot")
ax.set_xlabel("X")
ax.set_ylabel("Y")

# Show the plot
plt.show()
```
